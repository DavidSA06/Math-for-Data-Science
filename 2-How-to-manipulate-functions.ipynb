## Dependencies

```
import numpy as np
import matplotlib.pyplot as plt
```

## Vertical and horizontal displacements 

Since $c$ is a constant greater than zero, then the graph:

- $y = f(x) + c$ shifts $c$ units up.
- $y = f(x) - c$ shifts $c$ units down.
- $y = f(x - c)$ shifts $c$ units to the right.
- $y = f(x + c)$ shifts $c$ units to the left.

```
N = 1000

def f(x):
  return x**2;

c = 10

x = np.linspace(-15, 15, num=N)

y = f(x) + c
#y = f(x) - c
#y = f(x + c)
#y = f(x - c)


fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()
ax.axhline(y = 0, color = 'r')
ax.axvline(x = 0, color = 'r')
```

## Lengthening and compressions

Since $c$ is a constant greater than zero, then the graph :

- $y = c \cdot f(x)$ lengthens the graph vertically by a factor of $c$.
- $y = \frac{1}{c} \cdot f(x)$ compresses the graph vertically by a factor of $c$.
- $y = f(c \cdot x)$ compresses the graph horizontally by a factor of $c$.
- $y =  f(\frac{1}{c} \cdot x )$ lengthens the graph horizontally by a factor of $c$.

```
N = 1000

def f(x):
  return np.sin(x);

c = 3

x = np.linspace(-15,15, num = N)

y = c * f(x)
#y = 1/c * f(x)
#y = f((1/1)*x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()
ax.axhline(y = 0, color = 'r')
ax.axvline(x = 0, color = 'r')
```

## Reflections

- $y = -f(x)$ reflects the graph regarding the x axis.
- $y = f(-x)$ reflects the graph regarding the y axis.

```
N = 1000

def f(x):
  return x**3;

x = np.linspace(-10, 10, num = N)

y = f(x)
#y = -f(x)
#y = f(-x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()
ax.axhline(y = 0, color = 'r')
ax.axvline(x = 0, color = 'r')
```
