## Dependencies

```
from matplotlib import cm # To manage colors
import numpy as np
import matplotlib.pyplot as plt
```

## 3 dimension graphs

```
fig, ax = plt.subplots(subplot_kw = {"projection" : "3d"})


resolution = 100

X = np.linspace(-4, 4, resolution)
Y = np.linspace(-4, 4, resolution)

X, Y = np.meshgrid(X, Y)

Z = np.sin(X) + 2*np.sin(Y) 

# Gr√°ph the surface
surface = ax.plot_surface(X, Y, Z, cmap = cm.jet,
                       linewidth = 0, antialiased = False)

fig.colorbar(surface)

```

## Contour lines

```
fig2,ax2 = plt.subplots()
level_map = np.linspace(np.min(Z), np.max(Z), resolution) 
cp = ax2.contour(X, Y, Z, levels = level_map,cmap = cm.jet)
#cp = ax2.contourf(X, Y, Z, levels = level_map,cmap = cm.cool)
fig2.colorbar(cp)
ax2.set_title('Contour lines')
plt.show()
```
