## Gradient descent

## Dependencies

```
from matplotlib import cm # To manage colors
import numpy as np
import matplotlib.pyplot as plt
```

### 3D graph of the cost function

```
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

def f(x,y):
  return x**2 + y**2;

resolution = 100

X = np.linspace(-4, 4, resolution)
Y = np.linspace(-4, 4, resolution)

X, Y = np.meshgrid(X, Y)  #Return coordinate matrices from coordinate vectors

Z = f(X,Y)  #Do a paraboloid

# Gr√°ficar la superficie
surface = ax.plot_surface(X, Y, Z, cmap = cm.jet,
                       linewidth = 0, antialiased = False) #cmap for color

fig.colorbar(surface) #Insert a colorbar at the right

```

## Gradient Descent

```
level_map = np.linspace(np.min(Z), np.max(Z), resolution) 
plt.contourf(X, Y, Z, levels = level_map, cmap = cm.jet)  #"contourf" for a color fill instead of "contour" lines 
plt.colorbar()
plt.title('Gradient Descent')

def derivate(_p, p):
  return  (f(_p[0], _p[1]) - f(p[0], p[1])) / h #Using the derivative definition copy with increment minus calculate point

p = np.random.rand(2) * 8 - 4 # generate two random values

plt.plot(p[0], p[1], 'o', c='k')  #Graph the first point in black

lr = 0.01 #learning rate
h = 0.01  #an increment

gradient = np.zeros(2)

for i in range(10000):
  for idx, value in enumerate(p): #get the index and value
    _p = np.copy(p) #Return an array copy of the given object

    _p[idx] = _p[idx] + h;  #increment copy

    dp = derivate(_p, p) #partial derivative

    gradient[idx] = dp  #save partial derivative

  p = p - lr * gradient #Points to graph

  if(i % 10 == 0):  #Graph every 10 points
    plt.plot(p[0], p[1],'o', c = 'r')

plt.plot(p[0], p[1], 'o', c = 'w')  #Graph the last point
plt.show()

print("The minimum point is: ", p)
```
